```python
#Kaggle data - https://www.kaggle.com/aaron7sun/stocknews

import pandas as pd
import re
```

```python
#Reading csv file
data=pd.read_csv('C:\\Users\\91721\\OneDrive\\NLP - ML & DL\\Stock price sentiment analysis\\Combined_News_DJIA\\Combined_News_DJIA.csv')
data.head()
```

```python
#Checking type of data
type(data)
```

```python
#Data Pre-processing
#Removing the additional html tags from data
data = data.replace('b\"|b\'|\\\\|\\\"', '', regex=True)
data.tail(2)
```

```python
#Data cleaning - Checking for null values
data.isna().sum()
```

```python
#Checking shape of data (dataframe)
data.shape
```

```python
#Dropping the rows with null values since there were only 3 rows having null values
data.dropna(inplace=True)
```

```python
#Checking the shape of data after dropping null values
data.shape
```

```python
#Splitting data into test and train - Date feature is used to split the data
train=data[data['Date']<'20150101']
test=data[data['Date']>'20141231']
```

```python
#Feature engineering - Removing punctuation and everything else except a-z and A-Z
data1=train.iloc[:,2:]
data1.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data1.head(2)
```

```python
#Renaming columns for ease of access
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data1.columns=new_index
data1.head(2)
```

```python
# Convertng headlines to lower case
for index in new_index:
    data1[index]=data1[index].str.lower()
data1.head(1)
```

```python
#Appending top reviews (combining text of all columns)
headline=[]
for i in range(len(data1.index)):
    headline.append(' '.join(str(x) for x in data1.iloc[i,0:25]))
```

```python
headline[0]
```

```python
#bag of words model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvectorizer=CountVectorizer(ngram_range=(2,2))
train_data1=countvectorizer.fit_transform(headline)
```

```python
#Random forest classifier
randomForestClassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomForestClassifier.fit(train_data1,train['Label'])
```

```python
#Transforming test data

test_transform=[]

for i in range(len(test.index)):
    test_transform.append(' '.join(str(j) for j in test.iloc[i,2:]))
test_dataset=countvectorizer.transform(test_transform)
prediction=randomForestClassifier.predict(test_dataset)
```

```python
#Importing libaries from sklearn package to check the performance of RF model
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#Confusion matrix
matrix=confusion_matrix(test['Label'],prediction)
print('Confusion matrix is : \n', matrix)
#Accuracy score
score=accuracy_score(test['Label'],prediction)
print('Accuracy score is : \n',score)
#Report
report=classification_report(test['Label'],prediction)
print('Classification report : \n',report)
```
